**Trimming Long Headers**

```python
import re

input_file = open('HBV_core_protein_170_215_21015.fasta','r')
a = open('HBV_core_protein_trim.fasta','w')
for line in input_file:
    if line.startswith(">"):
        header = re.search("\w+.\d+", line).group()
        a.write(">" + header + "\n")
    else:
        a.write(line)
        
input_file.close()
a.close()
```

**Length Criteria** Filters the sequences based on length criteria

```python
seq = ''

with open("HBV_core_protein_90.fasta", "r") as input_file, \
   open("HBV_core_protein_len_range.fasta", "w") as a, \
   open("HBV_core_protein_seq_length.txt", "w") as b:
        for line in input_file:
            if line.startswith(">"):
                if seq != '':
                    length = len(seq)
                    b.write(header + seq + "\n" + "length = " + str(length) + "\n")
                    if 180 <= length <= 185:
                        a.write(header + seq + "\n")
                seq = ''
                header = line
            else:
                seq += line.rstrip()
                
        if seq != '':
            length = len(seq)
            b.write(header + seq + "\n" + "length = " + str(length) + "\n")
            if 180 <= length <= 185:
                a.write(header + seq + "\n")
    
a.close()
b.close()
input_file.close()
print("Finish")
```

**Removing Ambiguous sequences**

```python
input_file = open ('HBV_core_protein_len_range.fasta', 'r')
a = open ('HBV_core_protein_pure_seq.fas', 'w') # output file name containing umambiguous sequences
b = open ('HBV_core_protein_ambi_seq.txt', 'w') # output file name containing ambiguous sequence headers
my_list = []

for line in input_file:
    if line.startswith(">"):
        header = line
        seq = input_file.readline().rstrip()
        seq = seq[:len(seq) - 1]
        length = len(seq)
        c = 0
        for i in range(length):
            if seq[i] not in 'ARNDCEQGHILKMFPSTWYVarndceqghilkmfpstwyv-':
                c = c + 1
                my_list.append(seq[i])
        if c == 0:
            a.write(header + str(seq) +'\n')
        else:
            b.write(header + str(c) + '\n' + str(my_list) + '\n')
            my_list = []
            
a.close()
b.close()
input_file.close()
print('Complete')
```

**Removing sequences with long gaps for MSA** ## All have long gaps

```python
with open('HBV_core_protein_pure_seq.fas', 'r') as f1, open('HBV_core_protein_msa.fasta', 'r') as f2, open('HBV_core_protein_gap_filter.fasta', 'w') as output:
    seq = ''
    seq_ori = ''
    sequences = []
    sequences_ori = []
    header = []
    gap_count = 0
    list_gaps = []
    
    for line1 in f2:
        if line1.startswith(">"):
            if seq != '':
                sequences.append(seq)
                seq = ''
            header.append(line1.rstrip())
        else:
            seq += line1.rstrip()
            
    for line2 in f1:    
        if line2.startswith(">"):
            if seq_ori != '':
                sequences_ori.append(seq_ori)
                seq_ori = ''
        else:
            seq_ori += line2.rstrip()

    sequences.append(seq)
    sequences_ori.append(seq_ori)
        
    for sequence in sequences:
        for ntd in range(len(sequence)):
            if sequence[ntd] == '-':
                gap_count += 1
            else:
                if gap_count > 0:
                    list_gaps.append(gap_count)
                    gap_count = 0

        if any(gap >= 10 for gap in list_gaps):
            list_gaps = []
            continue
        else:
            position = sequences.index(sequence)
            output.write(header[position] + "\n" + sequences_ori[position] + "\n")
            list_gaps = []
```

**Making Substitution Matrix**

```python
from Bio import AlignIO
import numpy as np
import pandas as pd
import math
from collections import Counter
import matplotlib.pyplot as plt
```

```python
filename = "HBV_core_protein_msa.fasta"
alignment = AlignIO.read(filename, "fasta")
print(alignment)
```

```python
total_counter = Counter()

for record in alignment:
    total_counter.update(record.seq)

total_counter
```

```python
col_labels = ['A', 'R', 'N', 'D', 'C', 'Q', 'E', 'G', 'H', 'I', 'L', 'K', 'M', 'F', 'P', 'S', 'T', 'W', 'Y', 'V']
```

```python
substitution_counts = {}

for i in range(alignment.get_alignment_length()):
    col = alignment[:, i]
    for j in range(len(col)):
        for k in range(j+1, len(col)):
            substitution_pair = (col[j], col[k]) # Creates a tuple
            substitution_counts[substitution_pair] = substitution_counts.get(substitution_pair, 0) + 1

# get method tries to retrieve the value associated with the key substitution_pair from the substitution_counts dictionary.
# If the key is not present, it returns a default value of 0
print(substitution_counts)
```

```python
substitution_matrix = pd.DataFrame(0, index = col_labels, columns = col_labels)
substitution_matrix
```

```python
for (aa1, aa2), count in substitution_counts.items():
    substitution_matrix.loc[aa1, aa2] = count

# For each pair of amino acids, this line sets the corresponding entry in the substitution_matrix DataFrame to the count.

substitution_matrix.to_excel('substitution_matrix_control.xlsx')
substitution_matrix
```

```python
accepted_pmm = pd.read_excel("substitution_matrix_control.xlsx", index_col = 0)
accepted_pmm
```

```python
accepted_pmm += accepted_pmm.transpose()
accepted_pmm /= 2
accepted_pmm
```

```python
A = 100 / ((accepted_pmm.loc[:,'A'].sum() - accepted_pmm.loc['A','A']) / total_counter['A'])
col_labels.remove('A')
```

```python
# Relative Mutabilities
relative_mutabilities = {}
relative_mutabilities['A'] = 100
for i in col_labels:
    relative_mutabilities[i] = round(((accepted_pmm.loc[:,i].sum() - accepted_pmm.loc[i, i]) / total_counter[i]) * A)
    
relative_mutabilities
```

```python
col_labels = ['A', 'R', 'N', 'D', 'C', 'Q', 'E', 'G', 'H', 'I', 'L', 'K', 'M', 'F', 'P', 'S', 'T', 'W', 'Y', 'V']
```

```python
# Calculating frequencies of all amino acids
freq = {}
for j in col_labels:
    freq[j] = (accepted_pmm.loc[:,j].sum() - accepted_pmm.loc[j,j]) / relative_mutabilities[j] # or dayhoff_relative_mutabilities
    
sum_freq = sum(freq.values())
for i, j in freq.items():
    freq[i] = round((j / sum_freq), 3)

freq
```

```python
# Calculating choice of proportionality lambda
s = 0
for i in col_labels:
    s += freq[i] * relative_mutabilities[i] # or dayhoff_relative_mutabilities
    
prop_lambda = (1 - 0.99) / s
prop_lambda
```

```python
# Creating the mutation matrix
mutation_matrix = pd.DataFrame()
for i in col_labels:
    for j in col_labels:
        mutation_matrix.loc[i,j] = (prop_lambda * relative_mutabilities[j] * accepted_pmm.loc[i,j]) / (accepted_pmm.loc[:,j].sum() - accepted_pmm.loc[j, j]) # or dayhoff_relative_mutabilities
        
mutation_matrix
```

```python
for i in col_labels:
    for j in col_labels:
        if i == j:
            mutation_matrix.loc[i,j] = 1 - (prop_lambda * relative_mutabilities[j]) # or dayhoff_relative_mutabilities
            
mutation_matrix
```

```python
pam_n = mutation_matrix.copy()
pam_1 = mutation_matrix.copy()

for _ in range(250):
    pam_n = pam_n.dot(pam_1)
    
# for i in col_labels:
#     for j in col_labels:
#         pam_n.loc[i,j] = round(pam_n.loc[i, j] * 100)
        
pam_n
```

```python
# Matrix without element rounded off

for i in col_labels:
    for j in col_labels:
        pam_n.loc[i, j] = (math.log10(pam_n.loc[i, j] / freq[i])) * 10
        
pam_n.to_excel('pam_250_control_wr.xlsx')
```

**Calculating Entropy at every sequence position**

The more conserved the column, the smaller its entropy.

```python
entropy = []

for i in range(alignment.get_alignment_length()):
    entropy_cal = 0
    freq = Counter()
    col = alignment[:, i]
    freq.update(col)
    for i in freq.values():
        entropy_cal += -(i * math.log2(i))
        
    entropy.append(entropy_cal)
    
plt.xlim(1, alignment.get_alignment_length())
plt.plot(entropy)
plt.xlabel("Sequence Position")
plt.ylabel("Entropy")
plt.title("Entropy for every position")
plt.show()
```

**Filter variable positions i.e. positions with high entropy values (Assumed cutoff: -46000)**

```python
print("The variable positions are:")
for i in entropy:
    if i > -46000:
        print(entropy.index(i))
```
